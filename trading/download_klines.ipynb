```python
import pandas as pd
from time import sleep
from trainer.data import download_history_from_symbols_list 
import logging
logging.basicConfig(format='%(asctime)s: %(message)s',
                    datefmt='%d/%m/%y %H:%M:%S', level=logging.INFO)
```

```python
kwargs = dict(
    symbols = ['BTCUSDT', 'DOGEBTC', 'DOGEUSDT'],
    tfs = ['1m', '5m', '15m', '30m', '1h', '4h', '1d', '1w']
)
download_history_from_symbols_list(**kwargs)
```

```python
min_last_date = None
for symbol in kwargs['symbols']:
    for tf in kwargs['tfs']:
        df = pd.read_csv(f'klines/{symbol}_{tf}.csv')
        df = df.rename({'open_time': 'date'}, axis=1)
        df['date'] = pd.to_datetime(df['date'], unit='ms')
        if min_last_date is None:
            min_last_date = df['date'].max()
        if df['date'].max() < min_last_date:
            min_last_date = df['date'].max()
min_last_date
```
