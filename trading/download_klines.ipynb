```python
from binance.um_futures import UMFutures
from binance.cm_futures import CMFutures
from binance.spot import Spot
import pandas as pd
from time import sleep

um_futures_client = UMFutures()
```

```python
def download_klines(file:str, symbol:str, timeframe='15m', count=1, client='um') -> None:
    cols = ['open_time', 'open', 'high', 'low', 'close', 'vol', 'close_time', 'qa_vol', 'trades',]
    if client == 'cm':
        _client = CMFutures()
    elif client == 'um':
        _client = UMFutures()
    else:
        _client = Spot()
    try:
        df = pd.read_csv(file)
    except:
        res = _client.klines(symbol, timeframe)
        df = pd.DataFrame(res).drop([9, 10, 11], axis=1)
        df.columns = cols
        df.to_csv(file, index=False)
        df = pd.read_csv(file)
    for i in range(count):
        min_timestamp = df['open_time'].min()-1
        res = _client.klines(symbol, timeframe, endTime=min_timestamp)
        _df = pd.DataFrame(res).drop([9, 10, 11], axis=1)
        _df.columns = cols
        df = pd.concat([_df, df], ignore_index=True)
        df.to_csv(file, index=False)
        print(f'Iteration {i+1} in {count}.')
```

```python
symbol = 'BTCUSDT'
tf = '1m'
```

```python
download_klines('klines/' + f'{symbol}_{tf}.csv', symbol, timeframe=tf, count=500, client='um')
```

```python
df = pd.read_csv('klines/' + f'{symbol}_{tf}.csv')
df['open_time'] = pd.to_datetime(df['open_time'], unit='ms')
```
